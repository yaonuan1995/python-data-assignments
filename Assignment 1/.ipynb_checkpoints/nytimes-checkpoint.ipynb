```python
import requests
import csv
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.nytimes.com/section/world')
```

```python
nyt_str = r.text
nyt_str
```

```python
data = BeautifulSoup(nyt_str,"html.parser")
```

```python
data
```

```python
headline = data.find_all('h2',attrs = {'class':"headline"})
headline[1].text.strip()
```

```python
nyt_headline = []
for i in range(len(headline)):
    a_headline = headline[i].text.strip()
#     print(a_tag)
    nyt_headline.append(a_headline)
nyt_headline
```
